```python
#Function Definition:
def my_function():
    # This is a block of code inside a function
    print("Hello, World!")
```

```python
#Conditional Statements (if, elif, else):
condition=True
if condition:
    # Block of code for if statement
    print("Condition is True")
elif another_condition:
    # Block of code for elif statement
    print("Another condition is True")
else:
    # Block of code for else statement
    print("Condition is False")
```

```python
#For Loop:
for i in range(5):
    # Block of code inside for loop
    print(i)
```

```python
#While Loop:
i=1
while i<11:
    print(i)
    i+=1
```

```python
#Class Definition:
class MyClass:
    # Block of code defining a class
    def __init__(self, value):
        self.value = value
    
    def display(self):
        print(self.value)
```

```python
#Try-Except Block:
try:
    x=True/False
    print(x)
except ZeroDivisionError as e:
    # Block of code to handle exceptions
    print("An error occurred:", e)
```

```python
# This is a single-line comment in Python
print("Hello, World!")  # This is an inline comment
```

In the example above, the comment # This is a single-line comment explains that the line is a comment, and the comment # This comment is next to a statement provides additional information about the print statement. Using comments effectively helps other developers (and yourself) understand the code better.

```python
import sys
print(sys.version_info)
```

```python
"""sys.version: This attribute also from the sys module gives a human-readable string representation of the Python version."""
import sys
print(sys.version)
```

```python
'''platform.python_version(): This function from the platform module returns a string representing the Python version in a human-readable format.'''
import platform
print(platform.python_version())
```

```python
#platform.python_version_tuple(): This function from the platform module returns the Python version as a tuple of strings, making it easier to perform version checks programmatically.
import platform
print(platform.python_version_tuple())
```

```python
# Syntax: lambda parameters: expression
add = lambda x, y: x + y

# Using the lambda function
result = add(3, 5)
print(result)  # Output: 8
```

In this example, the lambda x, y: x + y defines an anonymous function that takes two parameters (x and y) and returns their sum. This lambda function is then assigned to the variable add, and it can be called like a regular function.
